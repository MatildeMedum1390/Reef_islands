```python
notebook_name = 'Reef_Islands_Polar_Charts'

# Install pandas to read dataframe
import pandas as pd

# Import plotly express to visualize polar chart
import plotly.express as px

# Create the first dataframe for the Lady Musgrave Manual data, by asking pandas to read Excel
df = pd.read_excel('LM_DEA_azimuth.xlsx')

# A preview on how the dataset for Lady Musgrave DEA looks like:
# Year = [2016, 2017, 2018, 2019, 2020, 2021, 2022]
# Azimuth = [0, 107.9693, 138.4131, 169,5142, 152.4551, 147.3204, 140.7575]
# Distance = [0, 1.054433, 2.980013, 2.957286, 4.806863, 13.68689, 6.156398]
# Marker = [7, 6, 5, 4, 3, 2, 1]

# Create polar scatterplot, setting 'Distance' as the radius and 'Azimuth' as theta (angles).
# Centroids are colored according to 'Year', and 'Markers' from 1-7 is set as centroids size.
# The horizontal axis range is set to 0-15
fig = px.scatter_polar(df, r="Distance", theta="Azimuth", color="Year", size="Marker", range_r=[0, 15], color_discrete_sequence=px.colors.sequential.Plasma_r)

# Figure is updated to exchange tickvals from degress to ticktext with cardinal directions.
fig.update_layout(polar=dict(angularaxis=dict(tickmode='array',tickvals=[0,45,90,135,180,225,270,315],ticktext=['N','NE','E','SE',"S",'SW','W','NW'],),),)

# View figure
fig.show()
# Follow same procedure with all other datasets

# Lady Musgrave Manual data
df = pd.read_excel('LM_Manual_azimuth.xlsx')
fig = px.scatter_polar(df, r="Distance", theta="Azimuth", color="Year", size="Marker", range_r=[0, 15], color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.update_layout(polar=dict(angularaxis=dict(tickmode='array',tickvals=[0,45,90,135,180,225,270,315],ticktext=['N','NE','E','SE',"S",'SW','W','NW'],),),)
fig.show()

# Masthead DEA data. Horizontal axis range is set to 0-13
df = pd.read_excel('Mast_DEA_azimuth.xlsx')
fig = px.scatter_polar(df, r="Distance", theta="Azimuth", color="Year", size="Marker", range_r=[0, 13], color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.update_layout(polar=dict(angularaxis=dict(tickmode='array',tickvals=[0,45,90,135,180,225,270,315],ticktext=['N','NE','E','SE',"S",'SW','W','NW'],),),)
fig.show()

# Masthead Manual data
df = pd.read_excel('Mast_Manual_azimuth.xlsx')
fig = px.scatter_polar(df, r="Distance", theta="Azimuth", color="Year", size="Marker", range_r=[0, 13], color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.update_layout(polar=dict(angularaxis=dict(tickmode='array',tickvals=[0,45,90,135,180,225,270,315],ticktext=['N','NE','E','SE',"S",'SW','W','NW'],),),)
fig.show()

# Michaelmas DEA data. Horizontal axis range is set to 0-26
df = pd.read_excel('Mich_DEA_azimuth.xlsx')
fig = px.scatter_polar(df, r="Distance", theta="Azimuth", color="Year", size="Marker", range_r=[0, 26], color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.update_layout(polar=dict(angularaxis=dict(tickmode='array',tickvals=[0,45,90,135,180,225,270,315],ticktext=['N','NE','E','SE',"S",'SW','W','NW'],),),)
fig.show()

# Michaelmas Manual data
df = pd.read_excel('Mich_Manual_azimuth.xlsx')
fig = px.scatter_polar(df, r="Distance", theta="Azimuth", color="Year", size="Marker", range_r=[0, 26], color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.update_layout(polar=dict(angularaxis=dict(tickmode='array',tickvals=[0,45,90,135,180,225,270,315],ticktext=['N','NE','E','SE',"S",'SW','W','NW'],),),)
fig.show()
```
